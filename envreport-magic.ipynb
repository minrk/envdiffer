# Using %envreport magic in IPython

envreport defines an `%envreport` magic.

To define the magic, load envreport as an IPython extension:

````python
%load_ext envreport
````

By default, `%envreport` will produce a markdown report on `sys.prefix`.
In most cases, this is the first thing on $PATH, but `%envreport` will make sure it is when it collects the report.

````python
%envreport
````

## JSON output

You can also produce JSON output, which JupyterLab displays nicely:

````python
%envreport -f json
````

## Output for specific paths

If you specify a specific prefix,
which will be injected on $PATH to produce a report on another prefix.

(**note:** the path will _not_ be 'fully activated.'
If you want to report on an active environment,
activate the environment in the shell, then run `envreport` with no arguments.)

````python
import sys  # noqa

!{sys.executable} -m venv /tmp/other-env
````

````python
%envreport /tmp/other-env
````

## plain output

If you want the raw output as plaintext, add `--plain`.
This is the default if e.g. you run the magic in an IPython terminal instead of a notebook.

````python
%envreport --plain
````
